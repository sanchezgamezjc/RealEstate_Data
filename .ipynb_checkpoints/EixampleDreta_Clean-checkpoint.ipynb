```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import plotly.express as px
from unidecode import unidecode

import numpy as np
from scipy import stats

```

## Importar datos

```python
url = "https://raw.githubusercontent.com/sanchezgamezjc/RealEstate_Data/main/scrap_DretaEixample_PisosCompra_2021-06-03.csv"

full_data = pd.read_csv(url, na_values='N.A.')
```

```python
db = full_data
db
```

# Limpieza Dataset

```python
db.info()
```

Realizando un primer análisis de los datos descargados del portal inmobiliario podemos sacar diversas conclusiones rápidas que nos ayudara a realizar las primeras operaciones:

a) Al trabajar únicamente con los datos de un barrio, esta columna no nos es de gran utilidad sin embargo podemos observar que del total de 1102 filas que contiene la BBDD únicamente contienen barrio 1095. Estas filas cuyo "barrio" es Nulo deberán ser las primeras en eliminarse.

b) Para realizar el análisis de mercado necesitaremos que ciertas columnas tengan valores. Aquellas columnas designadas como necesarias son: "precio", "superficie" y "hab", y en ellas se deberá prescindir de todas aquellas entradas que no tengan valor, sean valor Nulo

c) En la descarga de las distintos viviendas se observa que muchas de ellas no son anuncios de viviendas, algunos anuncios se publicitan promociones completas u otro tipo de anuncios que no podemos tipificar para el estudio general. Todos aquellos elementos que en la columna "precio" contengan distintos valores que no sean precios deberán eliminarse.

#### a) Eliminar entradas donde columna barrio es Nulo (NaN)

```python
db = db[db['barrio'].notna()]

db.head()
```

#### b) Eliminaremos todas aquellas entradas que no tengan valores en las columnas designadas como necesarias (superficie, precio y habitación)

```python
db = db[db['superficie'].notna() & db['precio'].notna() & db['hab'].notna()]

db.head()
```

#### c) Eliminar aquellos datos que no son estrictamente viviendas.

```python
drop_price = db[(db['precio'] == 'desde') 
                | (db['precio'] == 'Oportunidad') 
                | (db['precio'] == 'Consultarprecio')
                | (db['precio'] == 'Consultar precio')]

db = db.drop(drop_price.index)

db.head()
```

#### d) Transformar tipo de columna

```python
db.dtypes
```

Según se aprecia en la celda anterior todas las columnas están consideradas de typo object, una tipologia especifica de python que indica mixticidad de valores dentro de las distintas columnas. Sin embargo para realizar el análisis de mercado nos interesa que ciertas columnas sean valores, estas columnas seran: superficie, planta, fecha_constr, precio, hab, baño, aseo y terraza.

a) En el caso de superficie, y terraza deberemos eliminar el texto que acompaña al valor para dejar unicamente el numero y eliminar cualquier espacio en blanco que quede antes o despues de dicho valor.

b) En las columnas de planta, fecha_constr, habitaciones, baños y aseo unicamente deberemos eliminar el texto que acompaña al valor.

c) Para finalizar en la columna de precio se debera eliminar el texto, en este caso el simbolo €, y eliminar el punto de los millares

```python
db['superficie'] = db['superficie'].str.strip('Superficie m')
db['superficie'] = db['superficie'].str.strip()

db['terraza'] = db['terraza'].str.strip('Terraza m / Balcón')
db['terraza'] = db['terraza'].str.strip()



db['planta'] = db['planta'].str.strip('Planta número')

db['fecha_constr'] = db['fecha_constr'].str.strip('Año construcción')

db['hab'] = db['hab'].str.strip('habitaciones')

db['baño'] = db['baño'].str.strip('Baños')

db['aseo'] = db['aseo'].str.strip('Aseo')



db['precio'] = db['precio'].str.strip('€')
db['precio'] = db['precio'].str.replace('.','', regex=False)
```

A continuación transformaremos las columnas seleccionadas a valores numericos para poder trabajar

```python
db[['superficie', 'planta', 'fecha_constr', 'precio', 'hab', 'baño', 'aseo', 'terraza']] = db[['superficie', 'planta', 'fecha_constr', 'precio', 'hab', 'baño', 'aseo', 'terraza']].apply(pd.to_numeric)
```

```python
db.info()
```

#### e) Limpieza outliers

Puede ocurrir que el anuncio creado desde la pagina web tenga datos erroneos. Estos datos erroneos tienen una repercusión elevada en los resultados, ya que frecuentemente son valores elevados, fuera de una distribución normal. No puede ser que una vivienda tenga 300.000 m² o que el precio sea de 320 millones. 

Estos valores son los llamados "outliers" y como criterio se eliminaran todos aquellos valores que esten por encima de 3 zscores.

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(15,15))

x = db['precio'] 
y = db['superficie']

fig.suptitle('Distribución valores con outliers')
ax1.scatter(x, y)
ax2.hist(x, 20)

plt.show()
```

```python
filt1 = np.abs(stats.zscore(db['precio'])) < 3
filt2 = np.abs(stats.zscore(db['superficie'])) < 3

filt = (filt1 & filt2)

dataset = db[filt]
```

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(15,15))

x = dataset['precio'] 
y = dataset['superficie']

fig.suptitle('Distribución valores sin outliers')
ax1.scatter(x, y)
ax2.hist(x, 20)

plt.show()
```

## Transformar datos

Para un correcto análisis de los datos es necesario transformar los valores obtenidos de distintas columnas. 

### a) Tratamiento valores nulos


En este caso es distintivo los valores obtenidos como Nulos en la gran mayoria de columnas. Estos valores Nulos nos indican falta de información cuyo valor no estaba incluido dentro del anuncio del portal inmobiliario. Para una mayor simplificación se trataran estos valores nulos no descargados como que esa muestra no incluye el elemento estudiado. 

Es decir, si el valor de parking de una de las muestras en Nulo en este caso se entendera que la muestra no incluye parking, en caso contrario el propietario lo hubiese especificado en el anuncio. Esto se repetira en las columnas de aseo, lavadero, terraza, trastero, ascensor, parking, acc y calefacción.

```python
db['aseo'] = db['aseo'].fillna(0)

db['lavadero'] = db['lavadero'].fillna('Sin lavadero')

db['terraza'] = db['terraza'].fillna(0)

db['ascensor'] = db['ascensor'].fillna('Sin ascensor')

db['trastero'] = db['trastero'].fillna('Sin trastero')

db['parking'] = db['parking'].fillna('Sin plaza parking')

db['acc'] = db['acc'].fillna('Sin aire acondicionado')

db['calefaccion'] = db['calefaccion'].fillna('Sin calefacción')
```

b) Tipificar los valores Lat y Lon

Es necesario limpiar los valore descargados para las columnas de Latitud (lat) y Longitud (lon). Nos quedaremos con los ultimos 10 caracteres de las muestras.

```python
db['lat'] = db['lat'].str[10:]

db['lon'] = db['lon'].str[10:]
```

#### b) Extraer indicador de reforma

Nos interes saber si la vivienda ha sido reformada. Como el portal inmobiliario no tiene una caracteristica específica a rellenar en su formulario por el anunciante, estos se ven obligados a indicarlo en la descripcion. Por lo tanto buscaremos un inicador de que la vivienda ha sida reformada mediante la busqueda de palabras clave, en este caso reforma.

```python
db['reforma'] = 'Sin reformar'
```

```python
filt_ref = db['text'].str.extract('(reforma)').notnull()[0]

db.loc[filt_ref, 'reforma'] = 'Reformado'
```

#### c) Complementar información planta

Se ha observado que distintos anuncios publicados en el portal, aunque el anunciante no indique el valor de la planta (altura de la vivenda en el edificio) en su ubicación especifica esta información aparece en la descripción general. Extraeremos este valor y actualizaremos el valor de la columan planta.

```python
def extraer_planta(x):

    
    x = str(x)
    
    x = x.lower()
    
    x = unidecode(x)
    
    
    
    if x == 'primera':
        return 1
        
    elif x == 'segunda':
        return 2
        
    elif x == 'tercera':
        return 3   
        
    elif x == 'cuarta':
        return 4
        
    elif x == 'quinta':
        return 5
        
    elif x == 'sexta':
        return 6
        
    elif x == 'septima':
        return 7
        
    elif x == 'octaba':
        return 8
    else:
        return None
    
```

```python
filt_plantatxt = db['planta'].isnull()

db['planta_txt'] = db.loc[filt_plantatxt,'text'].str.extract('(\w+)(?=\s+planta)')[0].apply(extraer_planta)



filt_planta = db['planta'].isnull()

db.loc[filt_planta,'planta'] = db['planta_txt']

db.drop(columns = 'planta_txt', inplace=True)
```

#### d) Incluir información duplex

Para realizar un análisis correcto se deberá asegurar que todos los valores sean lo mas homogeneos posibles. Observando distintas muestras se han encontrado distintas viviendas que corresponden a duplex. Por lo tanto debera crearse una columna nueva e indicar este valor.

```python
db['duplex'] = 'No'

filt_duplex = db['text'].str.extract('(duplex)').notnull()[0]

db.loc[filt_duplex, 'duplex'] = 'No'
```

#### e) Eliminar errores en introducción años

```python
filt_yr = db['fecha_constr'] > 1800
filt_nan = db['fecha_constr'].isnull()

db = db[filt_nan | filt_yr]
```
